# Feature Scaling Methods

## 1. Data with no outliers

```python
import numpy as np
import seaborn as sns
from math import pi
import matplotlib.pyplot as plt
```

```python
# generating 'moment' points
t = np.linspace(0, 2 * pi, 100)
```

```python
# plot points inside an elipse
c_x = 15
radius_x = 5

x = c_x + (radius_x * np.cos(t))
x
```

```python
# plot points inside an elipse
c_y = 2.5
radius_y = 1

y = c_y + (radius_y * np.sin(t))
y
```

```python
def find_fig_size(x, y, max_inches=10):
    span_w = x.max() - x.min()
    span_h = y.max() - y.min()

    ratio_wh = span_w / span_h

    # x-span is greater or equal to y-span
    if ratio_wh >= 1.0:
        fig_width = max_inches
        fig_height = max_inches / ratio_wh
    else:
        fig_width = max_inches * ratio_wh
        fig_height = max_inches

    return (fig_width, fig_height)
```

```python
plt.figure(figsize=find_fig_size(x, y, 10))
sns.scatterplot(x, y, color='gray', edgecolor='black', s=100)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
display()
plt.savefig('./images/data.png', bbox_inches='tight',dpi=300)
```

```python
data = np.hstack((x.reshape(x.size, 1), y.reshape(y.size, 1)))
data.shape
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
data_norm = scaler.fit_transform(data)
```

```python
plt.figure(figsize=find_fig_size(data_norm[:,0], data_norm[:,1], 10))
sns.scatterplot(data_norm[:,0], data_norm[:,1], s=100, color='#a0def9', edgecolor='black')
plt.axvline(0, 0, 1, c='gray', linestyle='dashed', linewidth=1)
plt.axhline(0, 0, 1, c='gray', linestyle='dashed', linewidth=1)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
display()
plt.savefig('./images/min-max-scaler-on-data.png', bbox_inches='tight', dpi=300)
```

```python
from sklearn.preprocessing import StandardScaler

std_scaler = StandardScaler()
data_std = std_scaler.fit_transform(data)

plt.figure(figsize=find_fig_size(data_std[:,0], data_std[:,1]))
sns.scatterplot(data_std[:,0], data_std[:,1], color='#fe7f87', edgecolor='black', s=100)
plt.axvline(0, 0, 1, c='gray', linestyle='dashed', linewidth=1)
plt.axhline(0, 0, 1, c='gray', linestyle='dashed', linewidth=1)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
display()
plt.savefig('./images/standard-scaler-on-data.png', bbox_inches='tight', dpi=300)
```

```python
from sklearn.preprocessing import RobustScaler

robust_scaler = RobustScaler()
data_robust = robust_scaler.fit_transform(data)

plt.figure(figsize=find_fig_size(data_robust[:,0], data_robust[:,1]))
sns.scatterplot(data_robust[:,0], data_robust[:,1], color='#FFDF9C', edgecolor='black', s=100)
plt.axvline(0, 0, 1, c='gray', linestyle='dashed', linewidth=1)
plt.axhline(0, 0, 1, c='gray', linestyle='dashed', linewidth=1)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
display()
plt.savefig('./images/robust-scaler-on-data.png', bbox_inches='tight', dpi=300)
```

```python
plt.figure(figsize=find_fig_size(data_std[:,0], data_std[:,1]))
sns.scatterplot(data_std[:,0], data_std[:,1], color='#fe7f87', edgecolor='black', s=100, label='standardization')
sns.scatterplot(data_norm[:,0], data_norm[:,1], s=100, color='#a0def9', edgecolor='black', label='normalization')
sns.scatterplot(data_robust[:,0], data_robust[:,1], color='#FFDF9C', edgecolor='black', s=100, label='robust scaler')
plt.axvline(0, 0, 1, c='gray', linestyle='dashed')
plt.axhline(0, 0, 1, c='gray', linestyle='dashed')
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.legend(fontsize=16)
display()
plt.savefig('./images/all-scalers-data.png', bbox_inches='tight', dpi=300)
```

# Data with outliers

```python
x_with_outliers = np.append(x, [0.8 * x.max(), 0.9 * x.max(), 0.85 * x.max(), 1 * x.max(), 1.05 * x.max()])
x_with_outliers
```

```python
y_with_outliers = np.append(y, [1.2 * y.max(), 2 * y.max(), 1.5 * y.max(), 1.8 * y.max(), 1.3 * y.max()])
y_with_outliers
```

```python
plt.figure(figsize=find_fig_size(x_with_outliers, y_with_outliers))
sns.scatterplot(x_with_outliers, y_with_outliers, color='gray', edgecolor='black', s=100)
plt.savefig('./images/data-wih-outliers.png', bbox_inches='tight', dpi=300)
```

```python
data_with_outliers = np.hstack((x_with_outliers.reshape(x_with_outliers.size, 1), y_with_outliers.reshape(y_with_outliers.size, 1)))
data_with_outliers.shape
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
data_norm = scaler.fit_transform(data_with_outliers)

plt.figure(figsize=find_fig_size(data_norm[:,0], data_norm[:,1]))
sns.scatterplot(data_norm[:,0], data_norm[:,1], s=100, color='#a0def9', edgecolor='black')
plt.axvline(0, 0, 1, c='gray', linestyle='dashed')
plt.axhline(0, 0, 1, c='gray', linestyle='dashed')
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
display()
plt.savefig('./images/min-max-scaler-on-data-with-outliers.png', bbox_inches='tight', dpi=300)
```

```python
from sklearn.preprocessing import StandardScaler

std_scaler = StandardScaler()
data_std = std_scaler.fit_transform(data_with_outliers)

plt.figure(figsize=find_fig_size(data_std[:,0], data_std[:,1]))
sns.scatterplot(data_std[:,0], data_std[:,1], color='#fe7f87', edgecolor='black', s=100)
plt.axvline(0, 0, 1, c='gray', linestyle='dashed')
plt.axhline(0, 0, 1, c='gray', linestyle='dashed')
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
display()
plt.savefig('./images/standard-scaler-on-data-with-outliers.png', bbox_inches='tight', dpi=300)
```

```python
from sklearn.preprocessing import RobustScaler

robust_scaler = RobustScaler()
data_robust = robust_scaler.fit_transform(data_with_outliers)

plt.figure(figsize=find_fig_size(data_robust[:,0], data_robust[:,1]))
sns.scatterplot(data_robust[:,0], data_robust[:,1], color='#FFDF9C', edgecolor='black', s=100)
plt.axvline(0, 0, 1, c='gray', linestyle='dashed', linewidth=1)
plt.axhline(0, 0, 1, c='gray', linestyle='dashed', linewidth=1)
plt.xticks(fontsize=20, rotation=60)
plt.yticks(fontsize=20)
display()
plt.savefig('./images/robust-scaler-on-data-with-outliers.png', bbox_inches='tight', dpi=300)
```

```python
plt.figure(figsize=find_fig_size(data_std[:,0], data_std[:,1]))
sns.scatterplot(data_std[:,0], data_std[:,1], color='#fe7f87', edgecolor='black', s=100, label='standardization')
sns.scatterplot(data_norm[:,0], data_norm[:,1], s=100, color='#a0def9', edgecolor='black', label='normalization')
sns.scatterplot(data_robust[:,0], data_robust[:,1], color='#FFDF9C', edgecolor='black', s=100, label='robust scaler')
plt.axvline(0, 0, 1, c='gray', linestyle='dashed')
plt.axhline(0, 0, 1, c='gray', linestyle='dashed')
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.legend(fontsize=16)
display()
plt.savefig('./images/all-scalers-data-with-outliers.png', bbox_inches='tight', dpi=300)
```
